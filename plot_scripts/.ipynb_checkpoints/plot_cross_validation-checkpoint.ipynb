```python
import numpy as np
import xarray as xr
import os 
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt 
import matplotlib.colors as clr 
import pandas as pd
import multiprocessing
import matplotlib.ticker as ticker
import scipy

plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

```python
os.chdir('/home/ayhwong/fs03_storage/software/GC_Emulator_zenodo/')
dpm25_gc_cross_validation = xr.open_dataarray('output/dpm25_gc_cross_validation.nc')
dpm25_gp_cross_validation = xr.open_dataarray('output/dpm25_gp_cross_validation.nc')
dpm25_gp_sigma_cross_validation = xr.open_dataarray('output/dpm25_gp_sigma_cross_validation.nc')
```

```python
# Do 2d histogram
dpm25_gc_cross_validation_1d = dpm25_gc_cross_validation.stack(dim=['Run','fold'])
dpm25_gp_cross_validation_1d = dpm25_gp_cross_validation.stack(dim=['Run','fold'])

dpm25_gp_cross_validation_1d = xr.where(dpm25_gp_cross_validation_1d == 0, 
                                        np.nan,dpm25_gp_cross_validation_1d)
```

```python
nas = np.logical_or(np.isnan(dpm25_gc_cross_validation_1d), 
                    np.isnan(dpm25_gp_cross_validation_1d))

dpm25_gc_cross_validation_1d_nonan = dpm25_gc_cross_validation_1d.values[~nas]
dpm25_gp_cross_validation_1d_nonan = dpm25_gp_cross_validation_1d.values[~nas]

mae = np.nanmean(np.abs(dpm25_gp_cross_validation_1d_nonan - dpm25_gc_cross_validation_1d_nonan))
corr = scipy.stats.pearsonr(dpm25_gp_cross_validation_1d_nonan, dpm25_gc_cross_validation_1d_nonan)[0]
mb = np.nanmean(dpm25_gp_cross_validation_1d_nonan - dpm25_gc_cross_validation_1d_nonan)

fig,ax = plt.subplots(figsize = (11,9))

im = ax.hist2d(dpm25_gc_cross_validation_1d_nonan,dpm25_gp_cross_validation_1d_nonan,bins = 100,norm=clr.LogNorm())
ax.axline((0,0),slope = 1, color = 'black' , lw =2, linestyle = '--')
cbar = fig.colorbar(im[3], ax=ax)

ax.set_xlabel(r'$\Delta$$PM_{2.5}$,GCHP',fontsize = 28, fontstyle = 'normal')
ax.tick_params(labelsize = 20)
ax.set_xlim(-60,80)
ax.set_ylim(-60,80)
ax.text(-50,60,r'$R^2$='+str(np.round(corr*corr,3)),fontsize = 24)
ax.text(-50,50,'MAE='+str(np.round(mae,3))+r' $\mu$g $m^{-3}$',fontsize = 24)
ax.text(-50,40,'MB='+str(np.round(mb,3))+r' $\mu$g $m^{-3}$',fontsize = 24)
ax.text(-50,30,'N='+str(len(dpm25_gc_cross_validation_1d_nonan)),fontsize = 24)

ax.set_ylabel(r'$\Delta$$PM_{2.5}$,GW-GPR',fontsize = 28)
cbar.set_label('Counts', rotation=270,fontsize = 28,labelpad = 30)
cbar.ax.tick_params(labelsize = 20)

#plt.savefig('Plots/fig_6.png',bbox_inches='tight',format = 'png',dpi = 300,facecolor ='white')
```
