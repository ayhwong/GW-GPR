```python
import numpy as np
import scipy
import xarray as xr
import os 
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt 
import matplotlib.colors as clr 
import pandas as pd

plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

```python
### load data

os.chdir('/home/ayhwong/fs03_storage/software/GC_Emulator_zenodo/')
st_pm25_normalized = xr.open_dataarray('output/st_pm25_all_vars.nc')
st_pm25_plot = xr.where(st_pm25_normalized != 0, st_pm25_normalized,np.nan)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3, figsize = (18,9), subplot_kw=dict(projection=ccrs.Robinson()))

subtitle = ['NO','BC','OC',r'$NH_{3}$',r'$SO_{2}$',r'$CO_{2}$','VOC','CH4','CO']
plt.subplots_adjust(hspace = 0.1,wspace = 0.05) 
for i,ax in enumerate(axes.flatten()):
    ax.set_global()
    ax.coastlines()
    ax.gridlines()
    im = st_pm25_plot[i,:,:].plot(ax= ax,transform = ccrs.PlateCarree(),cmap = 'Spectral_r',add_colorbar = False,vmin = 0, vmax = 1)
    ax.set_title(subtitle[i],fontsize=18)

cbar = fig.colorbar(im,ax=axes.ravel().tolist(),shrink = 0.9)
cbar.ax.tick_params(labelsize=14)

plt.savefig('Plots/fig_8.png',bbox_inches='tight',format = 'png',dpi = 300,facecolor ='white')
```
